```python
import matplotlib.pyplot as plt
```

#### 1. Use matplotlib to plot the following equation:
y = (x ** 2) - x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(11))
y0 = [n ** 2 - n + 2 for n in x]

plt.annotate('Origin', xy=(0, 0), xytext=(1, 25),
             arrowprops={'facecolor': 'blue'})

plt.plot(x, y0)
plt.show()
```

#### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
- y = x ** 1/2
- y = x ** 3
- y = 2 ** x

You can use functions from the math module to help implement some of the equations above.

```python
x = list(range(11))
y1 = [n ** (1/2) for n in x]
plt.plot(x, y1)
plt.show()
```

```python
y2 = [n ** 3 for n in x]
plt.plot(x, y2)
plt.show()
```

```python
y3 = [2 ** n for n in x]
plt.plot(x, y3)
plt.title('y = $2^x $')
plt.show()
```

#### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.subplot(2, 2, 1)
plt.plot(x, y0)

plt.subplot(2, 2, 2)
plt.plot(x, y1)

plt.subplot(2, 2, 3)
plt.plot(x, y2)

plt.subplot(2, 2, 4)
plt.plot(x, y3)
```

#### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x = range(11)
y0 = [n ** 2 - n + 2 for n in x]
y1 = [n ** (1/2) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, y0, c='blue')
plt.scatter(x, y1, c='orange')
plt.scatter(x, y2, c='green')
plt.scatter(x, y3, c='red')
plt.plot(x, y0, c='blue', label='x$^2$-x+2')
plt.plot(x, y1, c='orange', label='$x^{1/2}$')
plt.plot(x, y2, c='green', label='$x^3$')
plt.plot(x, y3, c='red', label='$2^x$')

plt.legend(loc='upper left')
plt.title('Combined Graph')
plt.xlabel('$x$')

plt.show()
```
