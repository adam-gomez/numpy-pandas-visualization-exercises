#### Title your chart "Big O Notation", Label your x axis "Elements", Label your y axis "Operations", Label your curves or make a legend for the curves, Use LaTex notation where possible

```python
import matplotlib.pyplot as plt
import math
```

#### Curves to graph
- y=0x+1 and label the curve "O(1)"

```python
x = list(range(1, 11))
y_01 = [0 * n + 1 for n in x]

plt.plot(x, y_01, label = 'O(1)')
plt.title('O(1)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = log(x) and label the curve "O(log n)"

```python
x = list(range(1, 11))
y_02 = [math.log(n,10) for n in x]

plt.plot(x, y_02, label = "O(log n)")
plt.title('O(log n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = x and label the curve "O(n)"

```python
x = list(range(1, 11))
y_03 = [n for n in x]

plt.plot(x, y_03, label = 'O(n)')
plt.title('O(n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = x * log(x) and label it "O(n log n)"

```python
x = list(range(1, 11))
y_04 = [n * math.log10(n) for n in x]

plt.plot(x, y_04, label = 'O(n log n)')
plt.title('O(n log n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = x ** 2 and label it "O(n^2)"

```python
x = list(range(1, 11))
y_05 = [n ** 2 for n in x]

plt.plot(x, y_05, label = 'O($n^2$)')
plt.title('O($n^2$)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = 2 ** x and label it "O(2^n)"

```python
x = list(range(1, 11))
y_06 = [2 ** n for n in x]

plt.plot(x, y_06, label = 'O($2^n$)')
plt.title('O($2^n$)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = x! and label it "O(n!)"

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

x = list(range(1, 6))
y_07 = [factorial(n) for n in x]

plt.plot(x, y_07, label = 'O(n!)')
plt.title('O(n!)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

- y = x ** x and label it "O(n^n)"

```python
x = list(range(1, 6))
y_08 = [n ** n for n in x]

plt.plot(x, y_08, label = 'O($n^n$)')
plt.title('O($n^n$)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```
